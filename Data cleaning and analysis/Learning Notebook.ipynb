## What are the monthly sales trends over the past year?

1. **January**: Solid sales performance, potentially driven by New Year's fitness resolutions. This could indicate strong alignment between product positioning and seasonal health goals.

2. **February**: Significant dip in sales, likely reflecting the post-holiday consumer fatigue and tightened discretionary spending after December splurges.

3. **MarchApril**: Sales remain flat or slightly depressed, suggesting a slow recovery period. No major school holidays or campaigns active here, so demand may be organic and low.

4. **MayJune**: Noticeable rise in sales aligns with mid-year school holidays. This pattern could point to increased family or youth purchases during time off, or better campaign alignment.

5. **JulyAugust**: Gradual decline likely reflects return to routine and school/work schedules. No notable campaigns or holidays appear to counteract this dip.

6. **September**: Spike in sales coincides with the third-term school holiday. This further strengthens the hypothesis that school break periods drive consumer activity.

7. **OctoberNovember**: Sales decline again, likely from post-holiday normalization. The lack of major campaigns during this window could also contribute to the softness.

8. **December**: As expected, sales rebound sharply in line with the holiday season, consistent with gifting and year-end consumer spending.

### Key Insight:
Sales appear to follow a **cyclical pattern** strongly tied to **school holiday calendars** and **year-end behavior**. Campaign activity appears to amplify these trends (to be verified in later analysis), but seasonality seems to be the primary driver. 

This insight could inform better campaign timing and inventory planning.

```python
import matplotlib.dates as mdates
import matplotlib.ticker as mtick
# Imports the libraries and gives them aliases. Nothing new here.

def plot_monthly_metric(df, campaigns_df, value_column, title, y_label): # Defining the new function.

# Plots a monthly metric (e.g., sales, AOV) with campaign shading.

# Parameters:
# df           : DataFrame with columns ['order_month', value_column]
# campaigns_df : DataFrame with columns ['campaign_id', 'start_date', 'end_date']
# value_column : Column name in df to plot on Y-axis
# title        : Chart title
# y_label      : Y-axis label

# The above are the variables in the new function.
    
    plt.figure(figsize=(14, 7)) # Defining the canvas size
    plt.plot(df['order_month'], df[value_column], marker='o', label=title) 
    # Plotting the graphs based on the variables in the custom function

    plt.title(title) # Titles the graph according to the variable
    plt.xlabel('Month') # Labels the x-axis with Month
    plt.ylabel(y_label) # Labels the y-axis with the variable
    plt.grid(True) #Adds a grid to the graph
    plt.xticks(rotation=45) # Rotates the y-axis labels 45 degrees for easy reading

    # Format Y-axis with commas
    plt.gca().yaxis.set_major_formatter(
        mtick.FuncFormatter(lambda x, _: f'{int(x):,}')
    )

    # Format X-axis as Month Year
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

    # Campaign shading
    num_campaigns = len(campaigns_df)
    colors = plt.colormaps.get_cmap('tab20').resampled(num_campaigns)

    for i, row in campaigns_df.iterrows():
        plt.axvspan(
            row['start_date'],
            row['end_date'],
            color=colors(i),
            alpha=0.3,
            label=f"Campaign {row['campaign_id']}"
            if f"Campaign {row['campaign_id']}" not in plt.gca().get_legend_handles_labels()[1]
            else None
        )

    # Remove duplicate legend entries
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), loc='upper left')

    plt.tight_layout()
    plt.show()
```

```python
# Let's Calculate first and last order dates, total spend, order count, and number of campaign orders per customer
# I need the first and last order dates to calculate the customer lifespan
# I need the customer lifespan to calculate order frequency
# Order frequency = order count / customer lifespan

customer_metrics = ( # Opens our nested function and gives it an alias
    orders.groupby('customer_id').agg( # Creates an aggregation function and groups by customer_id 
        # Performs calculations for distinct customer IDs
        first_order=('order_date', 'min'), # Calculates the first order by referencing the earliest (min) order date
        last_order=('order_date', 'max'), # Calculates the last order by referencing the oldest (max) order date
        total_spent=('total_sale', 'sum'), # Sums up all the sales values for each customer
        order_count=('order_id', 'nunique'), # Counts the number of orders for each customer
        campaign_orders=('campaign_id', lambda x: x.notna().sum())  # Counts the orders linked to campaigns
    )
    .reset_index() #Transforms it into a nice little dataframe for easy reference (Quickly becoming my favorite tool.)
)

customer_metrics['lifetime_days'] = (customer_metrics['last_order'] - customer_metrics['first_order']).dt.days + 1
# Calculates our customer lifetime in days (plus 1 to avoid division by zero if same day). Will be used later.

customer_metrics['AOV'] = customer_metrics['total_spent'] / customer_metrics['order_count'] # Calculates avg order value. 
customer_metrics['order_frequency'] = customer_metrics['order_count'] / customer_metrics['lifetime_days'] 
# Calculates order frequency

customer_metrics['campaign_attribution'] = (customer_metrics['campaign_orders'] / customer_metrics['order_count'])*100
# Calculates how many orders are linked to a campaign as a percentage
```
